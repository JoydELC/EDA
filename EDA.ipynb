## **Introduction to exploratory data analysis using seaborn** 



* Exploratory data analysis provides knowledge about the data.
* It is an important step in the development workflow of any artificial intelligence algorithm.

* It is useful especially when predictive models are planned to be developed.

* **Seaborn** is a graphing library and is supported in **matplolib** and **numpy**.



```python
# We import the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Example 1: Car Price Dataset

```python
df = pd.read_csv("data_price_cars.csv")
# To display the top 5 rows 
df.head(5) 
```

```python
df.tail(5)                        # To display the botton 5 rows
```

```python
df.dtypes # Type of features
```

## Remove irrelevant columns (Features)

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis=1)
df.head(5)
```

## Rename columns (Features)

```python
df = df.rename(columns={"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price" })
df.head(5)
```

## Remove duplicated rows (samples)

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df.count()      # Used to count the number of rows
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

## Remove missing or null values

```python
print(df.isnull().sum())
```

```python
df = df.dropna()    # Dropping the missing values.
df.count()
```

```python
print(df.isnull().sum())   # After dropping the values
```

## Detection of outliers

```python
sns.boxplot(x=df['Price'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

## Column Ratio (Features)

## Histogram

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

## Heat Maps

```python
plt.figure(figsize=(10,5))
c = df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

## Scatterplot

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['Price'])
ax.set_xlabel('HP')
ax.set_ylabel('Price')
plt.show()
```

## General statistics

```python
print(iris.describe())
```

## Variables analysis (Features)

## Probability Distribution

```python
for i, feature in enumerate(list(iris.columns)[:-1]):
    fg = sns.FacetGrid(iris, hue='Species', height=5)
    fg.map(sns.distplot, feature).add_legend()
    plt.show()
```

## Box Plots 

```python
sns.boxplot(x='Species',y='PetalLengthCm', data=iris) 
plt.show()
```

## Violin Plots

```python
sns.violinplot(x='Species',y='PetalLengthCm', data=iris, size=8)
plt.show()
```

## Scatter Plots

```python
iris.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm')
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(iris, hue="Species", height=7) \
   .map(plt.scatter, "PetalLengthCm", "PetalWidthCm") \
   .add_legend();
plt.show();
```

## Pair Plots

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(iris, hue="Species", height=3);
plt.show()
```
